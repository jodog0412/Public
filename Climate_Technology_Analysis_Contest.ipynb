# Global Climate Bilateral Development Visualization
해당 파일은 다음과 같은 목적으로 제작되었습니다.
- 2011년부터 2020년까지의 세계 GDP 대비 기후기술에 대한 상호개발(multilateral development)의 비율 추이를 시각화하였습니다.
- 기후기술에 대한 주요 국가의 상호 특허(bilateral patent)의 개수를 networkx 라이브러리를 이용하여 시각화하였습니다.  

해당 시각화 자료를 통해 기후기술에 대한 국가들의 상호 개발 및 협력 수준을 효과적으로 파악할 수 있습니다.

## Importing library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx
import warnings
warnings.simplefilter("ignore")
```

## R&D/GDP
첫 번째 데이터는 2011년부터 2020년까지의 년도, 기후기술에 대한 상호개발 비용, 세계 GDP, 총 3 유형의 데이터로 구성되어 있습니다.  
이를 바탕으로, seaborn 라이브러리를 이용하여 년도에 따른 R&D/GDP 수치의 추이를 파악하고자 하였습니다.

### 데이터 시각화

```python
df=pd.read_excel("./data/global_climate_finance_flow_for_GDP.xlsx",sheet_name=0, engine='openpyxl')

#열 추가 및 데이터 전처리
#multilateral development의 단위는 billion $, GDP의 단위는 trillion $입니다.
#development flow/GDP의 수치는 %로 나타낼 것이기 때문에 단위 조정을 위해 0.1을 곱하였습니다.
df['Climate R&D/GDP']=df['climate finance flow of the multilateral development']/df['GDP']*0.1
df_reduced=df.dropna(subset=['Climate R&D/GDP'])

#시각화
sns.set_style('whitegrid')
plt.figure(figsize=(9,6))
plt.title('Global Climate bilateral development/GDP from 2011 to 2020')
sns.barplot(data=df_reduced,x="year",y='Climate R&D/GDP')
plt.ylabel('bilateral development/GDP(%)')
plt.show()
```

### 시각화 자료 분석

2011년부터 2020년까지 bilateral development/GDP는 성장세를 이룩하였으나, 0.1%에 미치지 못하였습니다.

## Patents-International collaboration in technology development(bilateral)
oecd에서 개발한 해당 데이터는 기후기술에 대한 주요 국가의 상호협력 정도를 파악하고자 제작되었습니다.  
해당 데이터는 행과 열이 국가 이름으로, 값은 특허 개수로 구성되어 있습니다.  
따라서, networkx 라이브러리를 바탕으로 이러한 상호협력 정도를 네트워크 그래프를 통해 시각화하는 것이 효과적이라 생각하였습니다.  
  
  아래 코드는 데이터를 처리하고 시각화하는 과정입니다.

### 데이터 처리

```python
#데이터 다운로드
g2 = nx.DiGraph()
df2_raw=pd.read_excel("./data/Patents-International collaboration in technology development(bilateral).xlsx",sheet_name=0, engine='openpyxl')

#데이터 전처리
df2_raw.drop([12],inplace=True)
df2_raw.drop([df2_raw.columns[7]],axis=1,inplace=True)
provider=list(df2_raw.iloc[6:16,0])
reciever=list(df2_raw.iloc[4,2:])

df2_weight=df2_raw.iloc[6:16,2:]
df2_weight.index=provider
df2_weight.columns=reciever
df2_weight.replace('..',0,inplace=True)
df2_weight=df2_weight[provider]
```

### 네트워크 시각화

```python
#노드 생성 및 디자인
#node_raw 리스트의 각 원소는 국가에 따른 GDP(billion $)를 나타냅니다.
node_raw=[17.42,3.23,27.29,38.88,51.49,16.51,28.79,214.33,142.80,1.20] 
nodesize=list(np.array(node_raw)*17)
g2.add_nodes_from(provider)
node_map = []
for i in node_raw:
    if i > 100:
        node_map.append('#2471A3')
    elif i>10:
        node_map.append('#7FB3D5')      
    else:
        node_map.append('#D4E6F1')

#엣지 생성 및 디자인
weight=[]
edge_map=[]
length=len(provider)
for i in range(0,length):
    for j in range(0,length):
        if (i!=j and df2_weight.iloc[i,j]!=0):
            weight.append(df2_weight.iloc[i,j]*0.05)
            g2.add_edge(provider[i],provider[j])
            if df2_weight.iloc[i,j]>100:
                edge_map.append('#2E86C1')
            elif df2_weight.iloc[i,j]>10:
                edge_map.append('#85C1E9')
            else:
                edge_map.append('#D6EAF8')

#시각화
plt.figure(figsize=(10,8))
plt.title('Climate bilateral patent in environmental tech by country in 2019')
nx.draw(g2, with_labels=True,  
        node_color=node_map, font_size=9, node_size=nodesize, alpha=0.9,
        width=weight, edge_color=edge_map)
plt.show()
```

### 시각화 자료 분석
자료에서 노드의 크기는 각 국가의 GDP를, 선의 굵기는 상호개발한 특허의 개수를 나타냅니다.  

자료를 보면 미국, 중국을 중심으로 기후기술 교류가 이루어지고 있으며, 모로코, 콜롬비아 같은 개발도상국은 이러한 상호교류에서 어느 정도 분리되어있음을 확인할 수 있습니다.  
  
  
또 한국의 경우, 미국, 중국, 일본과 같은 주요 국가와는 상호협력이 어느 정도 이루어지고 있으나, 서유럽 선진국들과의 협력은 이에 비해 부족함을 확인할 수 있습니다.
