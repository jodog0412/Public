# Environment R&D Growth Rate per GDPR

```python
df=pd.read_excel("./data/energy_investment_status.xlsx",sheet_name=0, engine='openpyxl')
df['Climate R&D Growth Rate/GDPR']=df['climate finance rate of the multilateral development']/df['GDPR']*100
df_reduced=df.dropna(subset=['Climate R&D Growth Rate/GDPR'])
sns.set_style('whitegrid')
plt.figure(figsize=(9,6))
plt.title('Climate R&D Growth Rate/GDPR')
plt.ylabel('R&D/GDP')
sns.barplot(data=df_reduced,x="year",y='Climate R&D Growth Rate/GDPR')

plt.show()
```
